```python
import pandas as pd
import matplotlib.pyplot as plt

from scrypy.plots import distplot
from scrypy.figure_cone_plot_choice_shares import cone_plot_choice_shares
from scrypy.plots import convergence_plot
```

```python
# Cell only for travis to execute for a small number of iterations saved in the results directory
# Import the stored results.
qoi_mean_params_change_mean_edu = pd.read_pickle("scrypy/results/qoi_mean_params_change_mean_edu_df.pkl")
qoi_mean_params_base_occ_shares = pd.read_pickle("scrypy/results/qoi_mean_params_base_occ_shares_df.pkl")
qoi_mean_params_policy_occ_shares = pd.read_pickle("scrypy/results/qoi_mean_params_policy_occ_shares_df.pkl")

mc_base_occ_shares = pd.read_pickle("scrypy/results/mc_base_occ_shares_df.pkl")
mc_policy_occ_shares = pd.read_pickle("scrypy/results/mc_policy_occ_shares_df.pkl")
mc_change_mean_edu = pd.read_pickle("scrypy/results/mc_change_mean_edu_df.pkl")
mc_input_parameters = pd.read_pickle("scrypy/results/mc_input_parameters_df.pkl")

cone_plot_choice_shares()

plt.style.use('scrypy/_mplstyle/uq.mplstyle') # wanna include that in funciton.

distplot(mc_change_mean_edu.values.T, r"$\Delta$ in mean years of education", save=True)
convergence_plot(
    mc_change_mean_edu.values.T, qoi_mean_params_change_mean_edu["change_mean_schooling"][0],
    r"$\Delta$ in mean years of education",
    absolute_deviation=False, save=True
    )

convergence_plot(
    mc_change_mean_edu.values.T, qoi_mean_params_change_mean_edu["change_mean_schooling"][0],
    r"$\Delta$ in mean years of education",
    absolute_deviation=True, save=True
    )
```
